```python
#getdummies is used to perform one hot encoding
df_encoded = pd.get_dummies(data, columns=['Country','Purchased','Gender'])
```

```python
# Perform label encoding for the "Country" and "Purchased" columns
# You can use the factorize method to automatically encode categorical values
data["Country"] = pd.factorize(data["Country"])[0]
data["Purchased"] = pd.factorize(data["Purchased"])[0]
data["Gender"] = pd.factorize(data["Gender"])[0]
# Printing the DataFrame with label-encoded columns
print(data)
```
